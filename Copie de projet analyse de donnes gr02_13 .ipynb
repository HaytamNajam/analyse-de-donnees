# Top 5 Tv-Shows

```python
tv_shows = data[data['type'] == 'TV Show']
```

### convertion des valeurs 'Duration' en nombre

```python
tv_shows['duration'] = tv_shows['duration'].str.extract('(\d+)', expand=False).astype(int)
```

```python
top_5_tv_shows = tv_shows.groupby('title')['duration'].sum().sort_values(ascending=False).head(5)
```

```python
plt.bar(top_5_tv_shows.index.tolist(), top_5_tv_shows.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Titre de la série')
plt.ylabel('Nombre total de saisons')
plt.title('Top 5 des séries télévisées avec le plus grand nombre de saisons')
plt.show()
```

# diagramme de similarité

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
```

```python
data['description'] = data['description'].fillna('')
```

```python
tfidf = TfidfVectorizer(stop_words='english')
text_features = ['title', 'director', 'cast', 'country', 'listed_in', 'description']
text_data = data[text_features].astype(str).apply(lambda x: ' '.join(x), axis=1)
```

```python
tfidf_matrix = tfidf.fit_transform(text_data)
similarity_matrix = cosine_similarity(tfidf_matrix[:100], tfidf_matrix[:100])
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix, annot=True, cmap='YlGnBu', xticklabels=False, yticklabels=False)
plt.title('Diagramme de similarité')
plt.show()
```

```python
content_index = 0
similar_contents = list(enumerate(similarity_matrix[content_index]))
similar_contents = sorted(similar_contents, key=lambda x: x[1], reverse=True)
similar_contents = [(data.iloc[i]['title'], similarity) for i, similarity in similar_contents if i != content_index]

for content, similarity in similar_contents:
    print(f"Titre: {content}, Similarité cosinus: {similarity}")
```

# Durée moyenne

```python

def compute_average_duration(data):
    durations = data['duration'].str.extract(r'(\d+)')
    durations = durations.astype(float)
    average_duration = np.mean(durations)
    return average_duration


average_duration = compute_average_duration(data)
print(f"Durée moyenne : {average_duration} minutes")
```

# Moyenne des années de sortie

```python
def calculate_mean(data, column):
    values = data[column].astype(float)
    mean = np.mean(values)
    return mean

# Exemple d'utilisation
mean_value = calculate_mean(data, 'release_year')
print(f"Moyenne des années de sortie : {mean_value}")


```
